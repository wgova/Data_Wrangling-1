#### Drop columns with NaN values

```python
df1 = df.dropna(axis='columns')
```

```python
df1.set_index('ID', inplace=True)
```

```python
ag_list = []
ex_list = []
ne_list = []
es_list = []
c_list = []

for i in df1.values:
    ag = 0
    ex = 0
    ne = 0
    es = 0
    c = 0
    for j in i:
        a = int(j[1])
        b = int(j[4])
        if a == 1:
            ex += b
        elif a == 2:
            ag += b
        elif a == 3:
            c += b
        elif a == 4:
            es +=b
        elif a == 5:
            ne += b
    ag_list.append(ag)
    ex_list.append(ex)
    ne_list.append(ne)
    es_list.append(es)
    c_list.append(c)

df1['Extraversion'] = ex_list
df1['Agreeableness'] = ag_list
df1['Openness to New Experiences'] = ne_list
df1['Emotional Stability'] = es_list
df1['Conscientiousness'] = c_list
```

#### Computing the means and std deviations of subscales

```python
extraversion_mean = df1['Extraversion'].mean()
extraversion_stdev = stdev(df1['Extraversion'])

agreeable_mean = df1['Agreeableness'].mean()
agreeable_stdev = stdev(df1['Agreeableness'])

emo_stab_mean = df1['Emotional Stability'].mean()
emo_stab_stdev = stdev(df1['Emotional Stability'])

consc_mean = df1['Conscientiousness'].mean()
consc_stdev = stdev(df1['Conscientiousness'])

new_exp_mean = df1['Openness to New Experiences'].mean()
new_exp_stdev = stdev(df1['Openness to New Experiences'])

```

```python
extra_high = (extraversion_mean + extraversion_stdev)/2
extra_low = (extraversion_mean - extraversion_stdev)/2

agree_high = (agreeable_mean + agreeable_stdev)/2
agree_low = (agreeable_mean - agreeable_stdev)/2

emo_stab_high = (emo_stab_mean + emo_stab_stdev)/2
emo_stab_low = (emo_stab_mean - emo_stab_stdev)/2

consc_high = (consc_mean + consc_stdev)/2
consc_low = (consc_mean - consc_stdev)/2

new_exp_high = (new_exp_mean + new_exp_stdev)/2
new_exp_low = (new_exp_mean - new_exp_stdev)/2
```

```python
#Creating functions that will generate interpretations for the subscales.
def extra_interpret(dataframe):
    result = []
    for i in dataframe:
        if i in ['Extraversion', 'yjtguitg']:
            for x in dataframe[i]:
                if x < extra_low:
                    result.append('low')
                elif x > extra_high:
                    result.append('high')
                else:
                    result.append('medium')
        else:
            pass
    return result

def emo_stab_interpret(dataframe):
    result = []
    for i in dataframe:
        if i in ['Emotional Stability']:
            for x in dataframe[i]:
                if x < emo_stab_low:
                    result.append('low')
                elif x > emo_stab_high:
                    result.append('high')
                else:
                    result.append('medium')
        else:
            pass
    return result        
       
def consc_interpret(dataframe):
    result = []
    for i in dataframe:
        if i in ['Conscientiousness']:
            for x in dataframe[i]:
                if x < consc_high:
                    result.append('low')
                elif x > consc_low:
                    result.append('high')
                else:
                    result.append('medium')
        else:
            pass
    return result                   
def new_exp_interpret(dataframe):
    result = []
    for i in dataframe:
        if i in ['Openness to New Experiences']:
            for x in dataframe[i]:
                if x < new_exp_low:
                    result.append('low')
                elif x > new_exp_high:
                    result.append('high')
                else:
                    result.append('medium')
        else:
            pass
    return result                   
            
def agree_interpret(dataframe):
    result = []
    for i in dataframe:
        if i in ['Agreeableness']:
            for x in dataframe[i]:
                if x < agree_low:
                    result.append('low')
                elif x > agree_high:
                    result.append('high')
                else:
                    result.append('medium')
        else:
            pass
    return result            
        
            
       
       
        
```

```python
df1['Extraversion Interpretation'] = extra_interpret(df1)  
df1['Emotional Stability Interpretation'] = emo_stab_interpret(df1)
df1['Conscientiousness Interpretation'] = consc_interpret(df1)
df1['Openness to New Experiences Interpretation'] = new_exp_interpret(df1)
df1['Agreeableness Interpretation'] = agree_interpret(df1)
```
